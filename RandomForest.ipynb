```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import random

data=pd.read_csv("I://西瓜数据集2.csv")

data.loc[[3,5]]
```

```python
def getEntropy(data,data_index_list):
    #print(data)
    total_num=len(data_index_list)
    true_num=len(data.loc[list(data.loc[data_index_list].好瓜==1)])
    false_num=len(data.loc[list(data.loc[data_index_list].好瓜==-1)])

    if true_num==0:
        true_res=0
    else:
        true_res=-(true_num/total_num)*math.log((true_num/total_num),2)
    if false_num==0:
        false_res=0
    else:
        false_res=-(false_num/total_num)*math.log((false_num/total_num),2)
    return true_res+false_res

def part(data,data_index_list,att_list,total_att_list):
    if len(att_list)==0:
        return
    data_ent=getEntropy(data,data_index_list)
    print(data_ent)
    att_en_list=[]
    for att in att_list:
        value_set=set(data.loc[data_index_list][att])
        total_en=0
        for value in value_set:
            #print(data[att]==value)
            tmp=data[data[att].isin([value])].index.to_list()
            #print(tmp)
            #print(data_index_list)
            value_index_list=list(set(data_index_list).intersection(set(tmp)))
            #print(value_index_list)
            value_ent=getEntropy(data,value_index_list)
            #print(value_ent)
            total_en+=len(value_index_list)/len(data_index_list)*value_ent
            #value_ent=getEntropy(data,data.loc[data_index_list][att])
        att_en=data_ent-total_en
        att_en_list.append(att_en)
    print(att_en_list)
    min_att_index=att_en_list.index(max(att_en_list))
    min_att=att_list[min_att_index]
    
    if max(att_en_list)==0:
        return
    value_set=set(data.loc[data_index_list][min_att])
    
    print("---------------------------")
    print(min_att,data_index_list)
    print("%%%%%%%%%%%%%%%%%%%%%%%%%%%")
    for value in value_set:
        tmp=data[data[min_att].isin([value])].index.to_list()
        value_index_list=list(set(data_index_list).intersection(set(tmp)))
        print(value,value_index_list)
    print("---------------------------")
    
    print(value_set)
    for value in value_set:
        tmp=data[data[min_att].isin([value])].index.to_list()
        value_index_list=list(set(data_index_list).intersection(set(tmp)))
        print(value_index_list)
        next_total_att_list=total_att_list.copy()
        next_total_att_list.remove(min_att)
        
        
        next_att_list=random.sample(next_total_att_list, int(math.log(len(next_total_att_list),2)))
        
        part(data,value_index_list,next_att_list,next_total_att_list)
        

        
data_index_list=[i for i in range(0,len(data))]
total_att_list=data.columns.to_list()
total_att_list.remove('好瓜')
att_list=random.sample(total_att_list, int(math.log(len(total_att_list),2)))
print(att_list)
part(data,data_index_list,att_list,total_att_list)
```
